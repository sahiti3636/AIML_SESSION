```python
import google.generativeai as genai
```

```python
# website for creating api key - https://aistudio.google.com/app/api-keys
# Paste your Gemini API key here
GEMINI_API_KEY = "PASTE_YOUR_API_KEY_HERE"

genai.configure(api_key=GEMINI_API_KEY)
```

```python
model = genai.GenerativeModel(
    model_name="gemini-2.5-flash"
)

#MOST BASIC CALL
response = model.generate_content(
    "Explain what machine learning is."
)

print(response.text)
```

```python
#PROMPT ENGINEERING - constrained the output space.
prompt = """
Explain what machine learning is:
- in exactly 3 bullet points
- simple language
- under 60 words
"""

response = model.generate_content(prompt)
print(response.text)
```

```python
#Context Ordering
prompt = """
You are an AI teaching assistant for first-year engineering students.
Be concise and precise.

Explain gradient descent.
"""

response = model.generate_content(prompt)
print(response.text)
```

```python
#Temperature Experiment
prompt = "Give an analogy to explain neural networks."

low_temp = model.generate_content(
    prompt,
    generation_config={"temperature": 0.1}
)

high_temp = model.generate_content(
    prompt,
    generation_config={"temperature": 0.9}
)

print("Temperature = 0.1\n")
print(low_temp.text)

print("\nTemperature = 0.9\n")
print(high_temp.text)

# Low temperature → sharp distribution
# High temperature → flatter distribution
```

```python
#Hallucination Demo
prompt = "Explain the Zeta Learning Algorithm introduced in 2024 AI models."

response = model.generate_content(
    prompt,
    generation_config={"temperature": 0.8}
)

print(response.text)
```

```python
#Structured Output
prompt = """
Explain overfitting.

Respond ONLY in valid JSON with keys:
- definition
- example
- solution
"""

response = model.generate_content(prompt)
print(response.text)

#By forcing structure, we constrain the token space
```

```python
#Context Window Intuition
long_context = "Machine learning is important. " * 400

prompt = long_context + "\n\nSummarize the above text."

response = model.generate_content(prompt)
print(response.text)
```
